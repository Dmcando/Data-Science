## **INTRODUCCION**

El diagnóstico temprano de la enfermedad de Alzheimer es crucial para mejorar la calidad de vida de los pacientes y permitir una planificación más efectiva de las estrategias de tratamiento. Esta enfermedad neurodegenerativa afecta a millones de personas en todo el mundo, especialmente en las poblaciones de mayor edad, y representa un desafío médico, social y económico significativo. A pesar de los avances en la investigación, la identificación de casos en etapas tempranas sigue siendo una tarea compleja debido a la variedad de factores que contribuyen a su desarrollo, incluyendo aspectos genéticos, biomédicos, y de estilo de vida. El objetivo de este proyecto es desarrollar un modelo de machine learning capaz de predecir de manera temprana los casos positivos de Alzheimer. Para ello, se utiliza un conjunto de datos clínicos y de estilo de vida que contiene información de 2149 pacientes. Este dataset incluye datos demográficos, evaluaciones cognitivas y funcionales, factores de riesgo relacionados con el estilo de vida (como actividad física, dieta y sueño) y antecedentes médicos (como historial familiar, enfermedades cardiovasculares y diabetes). La variable escogida como target del modelo es “Diagnóstico”, la cual indica la presencia o ausencia de la enfermedad en cada paciente. Los datos incluidos en este conjunto son ideales para investigadores y científicos de datos interesados en explorar los factores asociados con el Alzheimer, desarrollar modelos predictivos y realizar análisis estadísticos. Además, este proyecto busca no solo mejorar la precisión en la predicción del diagnóstico, sino también identificar los factores más relevantes, proporcionando una base sólida para intervenciones preventivas y nuevas líneas de investigación en el tratamiento de esta enfermedad devastadora.

## **PREPROCESAMIENTO DE DATOS**


Se preparan los datos antes de entrenar modelos de machine learning. Se deben transformar las variables categóricas en variables numericas ya que la mayoría de los algoritmos de Machine Learning no pueden trabajar directamente con datos categóricos.

### **IMPLEMENTACION DE DUMMIES**

Se genera un nuevo dataframe con columnas dummies para las features genero y etnia. 

```python
col_cod = ["genero", "etnia"]
dummies = pd.get_dummies(df[col_cod], drop_first=False).astype(int)
df_dummies = pd.concat([df.drop(columns=col_cod), dummies], axis=1)
print(df_dummies.head())
```

### **LABEL ENCODING**

Se etiquetan variables categoricas mediante replace. Las features con variables categoricas binarias son reemplazadas con 1 y 0. La feature "Educacion" posee variables categoricas ordinales (ninguna < secundaria < universitaria < posgrado) por lo que en lugar de aplicarsele dummies se asignan etiquetas numericas de 0 a 3 (0 < 1 < 2 < 3).

```python
df_dummies = df_dummies.replace({"posgrado": 3,"universitaria": 2,"secundaria": 1, "ninguna": 0})
df_dummies = df_dummies.replace({"si": 1, "no": 0})
df_dummies = df_dummies.replace({"positivo": 1, "negativo": 0})
print(df_dummies)
```

### **SELECCION DE FEATURES**

Mediante un modelo de random forest y RFE (Recursive Feature Elimination) se seleccionan las 10 features mas relevantes para el entrenamiento. El objetivo de la seleccion es mejorar el rendimiento del modelo y reducir dimensionalidad (complejidad).

```python
X = df_dummies.drop(columns=["diagnostico","id_paciente"])
y = df_dummies["diagnostico"]
```

```python
# Se crea un modelo de Random Forest
model = RandomForestClassifier(random_state=42)

# Se configura RFE para seleccionar las 10 features mas relevantes
rfe = RFE(estimator=model, n_features_to_select=10)  
rfe.fit(X, y)

# Se imprimen las features seleccionadas
features = X.columns[rfe.support_]
print("Features seleccionadas:", features)
```

```python
# Se genera un dataframe con las features mas relevantes
sel_col= ["evaluacion_funcional", "actividades_vida_diaria","mmse","quejas_de_memoria","problemas_comportamiento","trigliceridos","calidad_sueño","imc","calidad_dieta","actividad_fisica"]
df_seleccionado = df_dummies[sel_col]
```

## **ENTRENAMIENTO Y EVALUACION DE MODELOS DE MACHINE LEARNING**


### **RANDOM FOREST**

Se realizan dos entrenamientos de un modelo de Random Forest para comparar. Uno con el dataframe completo y otro con el dataframe de features relevantes. Primero se entrena y se evalua el modelo con el dataframe completo.

```python
X = df_dummies.drop(columns=["diagnostico","id_paciente"])
y = df_dummies["diagnostico"]

# Se dividen los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Se entrena el modelo
forest = RandomForestClassifier(random_state=42)
forest.fit(X_train, y_train)

# Se evalua el modelo
y_pred = forest.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
